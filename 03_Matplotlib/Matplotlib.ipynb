### Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> Quickly
* OO API -> Advanced

#### Pyplot API

```python
x = np.linspace(0, 10, 1000)
x
```

```python
plt.plot(x, np.sin(x), color = "blue", linestyle = "dashed", label = "sin(x)");
plt.plot(x, np.cos(x), color = "red", label = "cos(x)")

plt.title("Sin(x) and Cos(x)")
plt.xlabel("X value")
plt.ylabel("F(X)")

# plt.xlim([0, 4]) #limited
# plt.xlim([-0.7, 0.6]) #limited
# plt.axis([0, 4, -0.7, 0.6]) # limited (xmin, xmax, ymin, ymax)
plt.axis("equal")

plt.legend(); # show caption label of graph
plt.show(); # in ipynb not necessary
```

#### OO API

```python
# 1. Prepared data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) #Figured size = Width & Height of plot

# 3. Plot the data
ax.plot(x, y);

ax.set(title = "A simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")
```

### Most common types of Matlotplib plots
* line
* scatter (phn tn)
* bar (ct)
* hist (tn sut)
* subplots()

#### Line

```python
# Create an array
x = np.linspace(0, 10, 100)
x[:5]
```

```python
## OO API plot a line chart 

fix, ax = plt.subplots() # fix = Figure, ax = Axes
ax.plot(x, x**3);
```

#### Scatter

```python
# Pyplot API
plt.scatter(x, np.exp(x)); # y = e^(x)
```

```python
# OO API

# Prepare new data
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000*rng.rand(100)

fig, ax = plt.subplots()
img1 = ax.scatter(x, y, s = sizes, c = colors, cmap='viridis', alpha = 0.6); # s:size, c: color, cmap:colormaps, alpha:opacity)

fig.colorbar(img1); 
```

#### Bar
* Vertical (ng)
* Horizontal (ngang)

```python
# Prepare data from a python dictionnary
soft_drink_prices = {"Coke" : 10,
                     "Pepsi" : 12,
                     "Sprite" : 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax.set(title = "Winmart's Soft Drink Prices", ylabel = "Price ($)"); 
```

```python
fig, ax =  plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values())); # parameters in barh use list value
```

#### Histogram

```python
# Prepare data

student_height = np.random.normal(170, 10, 250) # (location, std, quantity)
```

```python
student_height[:10], len(student_height)
```

```python
# Pyplot API
plt.hist(student_height, bins = 30);
```

#### Subplots

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# Plot date to each axis
ax1.plot(x, x/2);
ax1.plot(x, x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins = 30);
```

```python
# Option 2:
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
# Plot date to each axis
ax[0 ,0].plot(x, x/2);
ax[0, 0].plot(x, x*2);

ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1, 1].hist(student_height, bins = 30);
```
